```python
import pandas as pd
import os
```

```python
foler_path = "data"
os.listdir(foler_path)
```

```python
all_files = [f for f in os.listdir(foler_path) if f.endswith('.xlsx')]

df_list = []
for file in all_files:
    file_path = os.path.join(foler_path,file)
    df = pd.read_excel(file_path)
    df_list.append(df)

merged_df = pd.concat(df_list,ignore_index=True)

merged_df.head()
```

```python
merged_df.drop_duplicates(inplace=True)

if 'Order Date' in merged_df.columns:
    merged_df['Order Date'] = pd.to_datetime(merged_df['Order Date'],errors='coerce')
    merged_df.dropna(subset=['Order Date'],inplace=True)
```

```python
# Product-wise Sales
product_sales = merged_df.groupby('Product Name')['Sales'].sum().reset_index()
product_sales.head()
```

```python
# Catergory_Salles
category_sales = merged_df.groupby('Category')['Sales'].sum().reset_index()
category_sales
```

```python
# Region-wise Sales
if 'Region' in merged_df.columns:
    region_sales = merged_df.groupby('Region')['Sales'].sum().reset_index()
region_sales
```

```python
# Final
output_path = "Final_Sales_Report.xlsx"

with pd.ExcelWriter(output_path, engine='openpyxl') as writer:
    merged_df.to_excel(writer, sheet_name='Merged Data', index=False)
    product_sales.to_excel(writer, sheet_name='Product Sales', index=False)
    category_sales.to_excel(writer, sheet_name='Category Sales', index=False)
    if 'Region' in merged_df.columns:
        region_sales.to_excel(writer, sheet_name='Region Sales', index=False)

output_path
```
